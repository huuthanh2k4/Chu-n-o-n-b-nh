```python
# chuyển đổi dữ liệu thành số (đổi kiểu dữ liệu cho gender,  family_history_with_overweight, FAVC, CAEC, SMOKE, SCC, CALC, MTRANS, NObeyesdad)
doi_gender = {'Female':0, 'Male':1}
doi_family_history_with_overweight = {'yes':1, 'no':0}
doi_FAVC = {'yes':1, 'no':0}
doi_CAEC = {'no': 0, 'Sometimes': 1, 'Frequently': 2, 'Always': 3}
doi_SMOKE = {'yes':1, 'no':0}
doi_SCC = {'yes':1, 'no':0}
doi_CALC = {'no':0, 'Sometimes':1, 'Frequently':2, 'Always':3}
doi_MTRANS = {'Public_Transportation':0, 'Automobile':1, 'Motorbike':2, 'Bike':3, 'Walking':4}
doi_NObeyesdad = {'Insufficient_Weight':0, 'Normal_Weight':1, 'Overweight_Level_I':2, 'Overweight_Level_II':3, 'Obesity_Type_I':4, 'Obesity_Type_II':5, 'Obesity_Type_III':6}

# Insufficient_Weight (0) – Thiếu cân

# Normal_Weight (1) – Cân nặng bình thường

# Overweight_Level_I (2) – Thừa cân cấp độ I

# Overweight_Level_II (3) – Thừa cân cấp độ II

# Obesity_Type_I (4) – Béo phì loại I

# Obesity_Type_II (5) – Béo phì loại II

# Obesity_Type_III (6) – Béo phì loại III (nghiêm trọng)


df_obesity['Gender'] = df_obesity['Gender'].map(doi_gender)
df_obesity['family_history_with_overweight'] = df_obesity['family_history_with_overweight'].map(doi_family_history_with_overweight)
df_obesity['FAVC'] = df_obesity['FAVC'].map(doi_FAVC)
df_obesity['CAEC'] = df_obesity['CAEC'].map(doi_CAEC)
df_obesity['SMOKE'] = df_obesity['SMOKE'].map(doi_SMOKE)
df_obesity['SCC'] = df_obesity['SCC'].map(doi_SCC)
df_obesity['CALC'] = df_obesity['CALC'].map(doi_CALC)
df_obesity['MTRANS'] = df_obesity['MTRANS'].map(doi_MTRANS)
df_obesity['NObeyesdad'] = df_obesity['NObeyesdad'].map(doi_NObeyesdad)

df_obesity.sample(10)
```

```python
# làm tròn tuổi thành số nguyên cho age, FCVC, NCP, FAF, TUE
df_obesity['Age'] = df_obesity['Age'].round(0).astype(int)
df_obesity['FCVC'] = df_obesity['FCVC'].round(0).astype(int)
df_obesity['NCP'] = df_obesity['NCP'].round(0).astype(int)
df_obesity['FAF'] = df_obesity['FAF'].round(0).astype(int)
df_obesity['TUE'] = df_obesity['TUE'].round(0).astype(int)

df_obesity.sample(10)
```

```python
# làm tròn chiều cao  và cân nặng sau 2 dấu phẩy cho Height, Weight
df_obesity['Height'] = df_obesity['Height'].round(2)
df_obesity['Weight'] = df_obesity['Weight'].round(2)
df_obesity['CH2O'] = df_obesity['CH2O'].round(2)

df_obesity.sample(10)
```

# Tập huấn luyện X và biến độc lập y

```python
# Tập huấn luyện x
x = df_obesity.drop({'id','NObeyesdad'}, axis=1)
x
```

```python
# biến độc lập y
y = df_obesity[['NObeyesdad']]
y
```

# Huấn luyện mô hình

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, 
                                                    y, 
                                                    test_size=0.25, 
                                                    random_state=63)
print('Kích thước tập huấn luyện:', x_train.shape, y_train.shape)
print('Kích thước tập kiểm tra:', x_test.shape, y_test.shape)
```

```python
x_train
```

```python
y_train
```

```python
# Chuẩn hóa dữ liệu cho mô hình
from sklearn.preprocessing import StandardScaler
Chuan_hoa = StandardScaler()
x_train = Chuan_hoa.fit_transform(x_train)
x_test = Chuan_hoa.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

NutriAI = DecisionTreeClassifier(criterion='gini',
                                 splitter='best',
                                 max_depth=9,
                                 random_state=15)
NutriAI.fit(x_train,y_train)
Do_chinh_xac = NutriAI.score(x_test,y_test)
print('Độ chính xác của mô hình trên tập kiểm tra:',round(Do_chinh_xac*100,2),'%')
NutriAI
```

```python
from sklearn.tree import DecisionTreeClassifier

NutriAI = DecisionTreeClassifier(criterion='entropy',
                                 splitter='best',
                                 max_depth=12,
                                 random_state=10)
NutriAI.fit(x_train,y_train)
Do_chinh_xac = NutriAI.score(x_test,y_test)
print('Độ chính xác của mô hình trên tập kiểm tra:',round(Do_chinh_xac*100,2),'%')
NutriAI
```

```python
# Tìm tham số tối ưu cho model cây quyết định
from sklearn.model_selection import GridSearchCV

Tham_so = {
    'criterion': ['gini', 'entropy'],
    'max_depth': range(3,21,2),
    'min_samples_split': range(1,11),
    'min_samples_leaf': range(1,11)
}

search_luoi = GridSearchCV(estimator=DecisionTreeClassifier(random_state=42), param_grid=Tham_so , cv=5)
search_luoi.fit(x_train, y_train)

print('Tham số tốt nhất:',search_luoi.best_params_)
print('Độ chính xác tốt nhất:',round(search_luoi.best_score_*100,2))
```

```python
NutriAI = search_luoi.best_estimator_
NutriAI
```

```python
# Lưu model 
import pickle
save_file = open('Model/NutriAI.sav', 'wb')
a = pickle.dump((NutriAI,Chuan_hoa), save_file)
if a==None:
    print("Model đã được lưu thành công")
else:
    print("Lưu model thất bại")
```
